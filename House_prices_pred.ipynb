```python
data =pd.get_dummies(data,drop_first=True,dtype='int')
```

```python
data.shape
```

```python
data.select_dtypes(include='object').columns
```

```python
X =data.drop(columns='SalePrice')
X.head()
```

```python
y = data['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

```python
X_train, X_test
```

## Building the model

### multiple linear regression


```python
from sklearn.linear_model import LinearRegression
```

```python
line = LinearRegression()
line.fit(X_train, y_train)
y_pred = line.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

### randomforest regression


```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor()
forest.fit(X_train,y_train)
y_pred = forest.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

XGboost Regression

```python
from xgboost import XGBRFRegressor
```

```python
xgb = XGBRFRegressor()
xgb.fit(X_train,y_train)
y_pred = xgb.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

## Hyper parameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params = { 
    'n_estimators': [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000],
    'max_features': ['sqrt', 'log2', None, 0.2, 0.6, 1],
    'max_depth': [2, 4, 6, 8, 10, 12, 14, 16, None],
    'bootstrap': [True],
    'max_samples': [0.5, 0.6, 0.7, 0.8, 0.9, 1],
    'min_samples_split': [2, 5, 10, 15, 20],
    'min_samples_leaf': [1, 2, 4, 6, 8, 10]
    }
```

```python
rand = RandomizedSearchCV(forest, param_distributions=params,n_iter=50, cv =5, verbose =2, n_jobs=-1, random_state=0)
rand.fit(X_train,y_train)
```

```python
rand.best_estimator_
```

```python
rand.best_params_
```

```python
forest = RandomForestRegressor( n_estimators = 600,
 min_samples_split= 2,
 min_samples_leaf= 1,
max_samples= 0.9,
 max_features= 0.2,
 max_depth= None,
 bootstrap= True)
forest.fit(X_train,y_train)
y_pred = forest.predict(X_test)
```
